```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import dlib
```

```python
my_image_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/myg.jpeg'
img_bgr = cv2.imread(my_image_path) # load image by openCV
img_show = img_bgr.copy() # store image for print
plt.imshow(img_bgr)
plt.show()
```

```python
img_rgb = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
detector_hog = dlib.get_frontal_face_detector()
dlib_rects = detector_hog(img_rgb,1) #(image, num of image pyramid)
print(dlib_rects) # the box of detected face area
```

```python
for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()
    
    
    cv2.rectangle(img_show,(l,t),(r,b),(0,255,0),4,lineType=cv2.LINE_AA)

img_show_rgb = cv2.cvtColor(img_show,cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
model_path = os.getenv('HOME')+'/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)
```

```python
list_landmarks = []

for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb,dlib_rect)
    list_points = list(map(lambda p : (p.x,p.y), points.parts()))
    list_landmarks.append(list_points)
    
print(len(list_landmarks[0]))
```

```python
print(list_landmarks)
```

```python
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_show, point, 7, (0, 255, 255), -1)

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
for dlib_rect,landmark in zip(dlib_rects,list_landmarks):
    print(landmark[30]) # landmark[30] = nose
    x = landmark[30][0]
    y = landmark[30][1]+dlib_rect.height()//2
    w = h = dlib_rect.width()
    print('(x,y) : (%d,%d)'%(x,y))
    print('(w,h) : (%d,%d)'%(w,h))
```

```python
sticker_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/cat-whiskers.png'
img_sticker = cv2.imread(sticker_path) # Load sticker image 
img_sticker = cv2.resize(img_sticker, (w,h))
print (img_sticker.shape)
```

```python
refined_x = x - w // 2
refined_y = y - h
print ('(x,y) : (%d,%d)'%(refined_x, refined_y))
```

```python
if refined_x < 0: 
    img_sticker = img_sticker[:, -refined_x:]
    refined_x = 0
if refined_y < 0:
    img_sticker = img_sticker[-refined_y:, :]
    refined_y = 0

print ('(x,y) : (%d,%d)'%(refined_x, refined_y))
```

```python
sticker_area = img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8)
```

```python
plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```
