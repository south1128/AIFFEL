### Activate Function
`특정 조건` 이 만족하면 `활성화` 된다 

Activation function 을 사용하는 이유 ?
- 딥러닝 모델의 표현력을 향상시켜주기 위해서 ( 모델의 representation capacity 또는 expressicity 를 향상 )
- how come ? = 선형 관계로 들어오는 input 은 layer를 통과하며 아무리 곱하고 더해도 선형관계를 유지한다. 선형 관계에 비선형성을 줌으로써 비선형 데이터를 표현 가능 

The motivation of Activate Function

perceptron : <br>
- 딥러닝 모델의 여러개의 층 중 하나의 층을 가져와 쪼갠다면 노드 즉 퍼셉트론, 퍼셉트론을 쌓으면 단층 퍼셉트론이 되고, 단층 퍼셉트론을 쌓으면 다층 퍼셉트론 즉 우리가 아는 모델이 된다. 
- 동물의 학습 방법을 모방



perceptron 의 구조 
- 신경세포의 가지돌기에 해당 하는 구조로  input 이 들어옴 
- 각 입력항에 곱해지는 가중치는 신경 세포들의 연결 부위에 형성된 시냅스의 발달 정도
- 각 가지돌기에서 시냅스를 거쳐서 오는 신호들을 bias 와 sum 하여 activatioin function 을 거쳐 output 출력 

Activation function (transfer function):  <br>
들어온 신호가 특정 임계점을 넘으면 출력, 넘지 못하면 무시

- Linear activation function(선형 활성화 함수)
- Non-Linear activation function(비선형 활성화 함수)



일반적으로 딥러닝에서는 비선형 함수를 사용 
- how come ? 
  - 선형 변환 :  `선형` 이라는 규칙을 지키며 V 공간 상의 벡터를 W 공간 상의 벡터로 변환
  - 선형으로 이루어진 퍼셉트론 구조 상 , 선형 활성화 함수를 통해 가중치를 매번 업데이트 하는 것에 의미가 없어짐
  - Linear Activation Function 을 사용한다면 노드의 개수를 아무리 많이 붙여도 결국 하나의 노드를 사용하는 것과 차이가 없다.
  - 그래서 딥러닝 모델에는 비선형 함수를 사용함
