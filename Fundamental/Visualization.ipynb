```python
from datetime import datetime
import pandas as pd
import os

csv_path = '/content/drive/MyDrive/AMZN.csv'
data = pd.read_csv(csv_path,index_col=0,parse_dates=True)
price = data['Close']
# Pandas series has the best data algorithm for drawing line graphs
# what exactly price=data['Close'] means ?

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
price.plot(ax=ax,style='black')
# using pandas plot to draw figure, but also using matplotlib subplot to take position
plt.ylim([1600,2200])
plt.xlim(['2019-05-01','2020-03-01'])
# set X and Y range

# annotation
important_data = [(datetime(2019,6,3),'Low Price'),(datetime(2020,2,19),'Peak Price')]
for d, label in important_data:
  ax.annotate(label,xy=(d,price.asof(d)+10), # set position for annotation
              xytext=(d,price.asof(d)+100), # set position for annotation text
              arrowprops=dict(facecolor='red')) # add arrow and decoration
  
plt.grid() # grid on in figure 
ax.set_title('StockPrice')

plt.show()
```

```python
# Variation of linestyle and marker options 

x = np.linspace(0,10,100)

plt.plot(x,x+0,linestyle='solid')
plt.plot(x,x+1,linestyle='dashed')
plt.plot(x,x+2,linestyle='dashdot')
plt.plot(x,x+3,linestyle='dotted')
plt.plot(x,x+0,'-g') #solid green
plt.plot(x,x+1,'--c') #dashed cyan
plt.plot(x,x+2,'-.k') #dasheddot black
plt.plot(x,x+3,':r') #dotted red
plt.plot(x,x+4,linestyle='-')
plt.plot(x,x+5,linestyle='--')
plt.plot(x,x+6,linestyle='-.')
plt.plot(x,x+7,linestyle=':')
```

```python
grouped = df['tip'].groupby(df['sex'])
# data name 'tip' group by sex - > save total, mean, size etc from group
grouped.mean()

# sex
# Male      3.089618
# Female    2.833448
# Name: tip, dtype: float64

grouped.size()

# sex
# Male      157
# Female     87
# Name: tip, dtype: int64
```

```python
import numpy as np
sex = dict(grouped.mean())
# change mean data type to dictionary


x = list(sex.keys()) # put keys as x 
y = list(sex.values()) # put values as y

import matplotlib.pyplot as plt

plt.bar(x=x,height=y) #create bar graph (x=keys,y=values)
plt.ylabel('tips[$]')
plt.title('Tip by Sex')
```

```python
# histogram : 
# data 
mu1, mu2, sigma = 100, 130, 15
x1 = mu1 + sigma*np.random.randn(10000)
x2 = mu2 + sigma*np.random.randn(10000)

# drawing axis 
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

#  
patches = ax1.hist(x1, bins=50, density=False) #bins x  50   .
patches = ax1.hist(x2, bins=50, density=False, alpha=0.5)
ax1.xaxis.set_ticks_position('bottom') # x    
ax1.yaxis.set_ticks_position('left') #y   

# ,  
plt.xlabel('Bins')
plt.ylabel('Number of Values in Bin')
ax1.set_title('Two Frequency Distributions')

# 
plt.show()
```
